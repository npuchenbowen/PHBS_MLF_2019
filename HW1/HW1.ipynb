```python
from scipy.io import arff
import pandas as pd
```

```python
data = arff.loadarff('4year.arff')
df = pd.DataFrame(data[0])
```

```python
df['bankruptcy'] = (df['class'] == b'1')
```

```python
df.head()
```

```python
df.describe()
```

```python
sum(df.bankruptcy == True)
```

```python
df2 = df[['Attr1','Attr2','Attr7','Attr10','bankruptcy']]
df2.columns = ['X1','X2','X7','X10','Bankruptcy']
```

```python
print(df2)
```

```python
from sklearn.preprocessing import Imputer
imr = Imputer(missing_values='NaN',strategy='mean',axis=0)
imr = imr.fit(df2.values)
imputed_data = imr.transform(df2.values)
print(imputed_data)
```

```python
df3 = pd.DataFrame(imputed_data)
df3.columns = ['X1','X2','X7','X10','Bankruptcy']
df3.describe()
```

```python
df4 = df3[(df3['Bankruptcy']==True)]
df4.describe()
```

```python
df5 = df3[(df3['Bankruptcy']==False)]
df5.describe()
```
